```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.count() # 결측치를 제거한 값 개수 
```

```python
titanic.dtypes
```

```python
titanic.sum(numeric_only=True)
titanic.mean(numeric_only=True)
titanic.std(numeric_only=True)
titanic.max(numeric_only=True)
titanic.min(numeric_only=True)
# numeric_only = True 는 결측값을 포함한 계산 결과 
```

```python
titanic[['age','fare','survived']].sum(axis='rows') # 행간 연산(세로로 더함)을 함 
```

```python
titanic[['age','fare','survived']].sum(axis='columns') # 열간 연산(가로로 더함)을 함
```

```python
titanic.nlargest(3,columns=['fare','age']) # fare , age 열을 기준으로 가장 큰 것 3개
titanic.nsmallest(3,columns=['age'])# fare , age 열을 기준으로 가장 작은 것 3개
```

```python
titanic.describe()
```

```python
titanic.sort_values('fare')
```

```python
titanic.sort_values(by=['fare','age','sex'],ascending=[False,True,True],na_position='last')
```

```python
titanic.sort_index(ascending=False) # index 기준 정렬
```

```python
titanic.sort_index(axis=1) # column index 기준 정렬
```

```python
titanic.sort_index(axis=1 , ascending=False)
```

```python
titanic['sex'].value_counts().sort_values(ascending=False)
titanic['age'].value_counts().sort_values(ascending=False)
titanic['class'].value_counts().sort_values(ascending=False)
titanic['alive'].value_counts().sort_values(ascending=False)
```
