```python
import pandas as pd
survey_raw_df = pd.read_csv('survey_results_public.csv',index_col='ResponseId')
schema_raw = pd.read_csv("survey_results_schema.csv", index_col="qname")["question"]
```

---

```python
survey_raw_df['DevType'] # DevType데이터 확인 
```

```python
temp_words = survey_raw_df['DevType'].dropna()
temp_words
```

```python
temp_words = temp_words.str.replace(';',' ').str.replace(',',' ')
# ;랑 ,를 공백으로 변경 
temp_words
```

```python
words = ' '.join(temp_words)#' ' 구분으로 단어를 다 합쳐줌  
words
```

```python
wc = WordCloud(collocation_threshold=int(1e6),
               width=800,height=400,
               background_color='lightgrey').generate(words)

plt.figure(figsize=(16,8))

plt.axis('off')

plt.grid(False)
plt.imshow(wc)
plt.show()
```

```python
wc.to_file('DevType.png')
```
