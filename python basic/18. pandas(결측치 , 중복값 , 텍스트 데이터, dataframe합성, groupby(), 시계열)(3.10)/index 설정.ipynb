- 인덱스 여러개

```python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5,4),2),
                   columns=[['A','A','B','B'],['C1','C2','C1','C2']])
df3 # 다중열 인덱스 사용
df3.columns.names = ['Cidx1','Cidx2']
df3 # 다중열 인덱스 이름을 지정 
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6,4),2),
                   columns=[['A','A','B','B'],
                            ['C','D','C','D']],
                   index=[['M','M','M','F','F','F'],
                         ['id_' + str(i +1) for i in range(3)] *2])
df4.columns.names = ['Cidx1' , 'Cidx2']
df4.index.names = ['Ridx1' , 'Ridx2']
df4
```

```python
df4.stack('Cidx1') # 인덱스가 반시계로 회전한 효과
```

```python
df4.unstack('Ridx2') #인덱스가 시계 방향으로 회전한 효과
```

```python
df3[('B','C1')] # 인덱싱 하는 방법
```

```python
df3.loc[0, ('B','C1')] # 0행 1반 index B 에 2번 index C1
df4.loc[:,('A','C')]
df4.loc[('M','id_1'):]
```

```python
df4.loc[('M',slice(None)),:] # M 전체범위
df4.loc[(slice(None),'id_1'),:] # id_1 전체범위
## 다중인덱스에서는 :를 사용할 수 없으므로 slice(None)을 사용 
```

```python
df6 = df4.swaplevel("Cidx1","Cidx2",1) # 인덱스 교환
df6
```

```python
df5 = df4.swaplevel("Ridx1","Ridx2")
df5
```
