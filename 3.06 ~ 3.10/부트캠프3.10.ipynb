- 결측치 제거

```python
employees = pd.read_csv('datas/employees.csv',parse_dates=['Start Date'])
employees.dropna() #행에 결측치를 하나라도 가지면 제거 
employees.dropna(how='all') # 결측치를 모두 가지면 제거 

employees.loc[employees.Gender.isnull()] # gender 결측치 조회
employees.dropna(subset = ['Gender']) # geender 결측치 있으면 제거 
```

```python
cond1 = employees['Start Date'].isnull()
cond2 = employees['Salary'].isnull()
print(f'condi1:{cond1.sum()},cond2:{cond2.sum()}') #2개씩 결측치 있음

employees[cond1]  # 2개 결측치 어딨는지 2,1000 번쨰
employees[cond2]  # 2개 결측치 어딨는지 0,1000 번째
employees.dropna(subset=['Start Date','Salary'])
# 2, 1000 , 0번쨰 행 제거

employees[employees.notnull().sum(axis=1) >=4]
employees.dropna(how='any',thresh=4) # 한행에 4개 이상의 데이터가 있으면 쓰겠다
```

- 중복 제거

```python
employees['Team'].head()
employees['Team'].duplicated(keep='last')
#앞에서 발견한건 True 뒤에서 발견한것 Flase 
first_one_in_team = ~employees['Team'].duplicated()
employees[first_one_in_team] # 첫번쨰 등장값만 True반환 

#method 사용 
employees.drop_duplicates() # 6개의 값이 모두 동일한 행을 저게함
employees.drop_duplicates(subset=['Team']) # 특정 컬럼에 대해 중복 배제
employees.drop_duplicates(subset=['Team'],keep='last') 
# 특정 컬럼에 대해 중복 배제(마지막에 발견딘 값만 남김)

employees.drop_duplicates(subset=['Gender','Team']) 
# 둘다 중복이 되면 지움 
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6,4),2),
                   columns=[['A','A','B','B'],
                            ['C','D','C','D']],
                   index=[['M','M','M','F','F','F'],
                         ['id_' + str(i +1) for i in range(3)] *2])
df4.columns.names = ['Cidx1' , 'Cidx2']
df4.index.names = ['Ridx1' , 'Ridx2']
df4
```

```python
df4.stack('Cidx1') # 인덱스가 반시계로 회전한 효과
```

```python
df4.unstack('Ridx2') #인덱스가 시계 방향으로 회전한 효과
```

### groupby()

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1':['A','A','B','B','A'],
    'key2' : ['one','two','one','two','one'],
    'data1' :[1,2,3,4,5],
    'data2' :[10,20,30,40,50]
})
df2
```

```python
groups = df2.groupby(df2.key1)
groups.groups
```

```python
groups.sum() # key1을 기준으로 동일한 값 더해준다
```

```python
df2.data1.groupby(df2.key1).sum()
```

```python
df2.groupby(df2.key1)['data2'].sum()
df2.groupby(df2.key1).sum()['data2']
```

```python
df2.data1.groupby([df2.key1,df2.key2]).sum()
```

### freq 인자 
- s: 초
- T: 분
- H: 시간
- D: 일(day)
- B: 주말이 아닌 평일
- W: 주(일요일)
- W-MON: 주(월요일)
- M: 각 달(month)의 마지막 날
- MS: 각 달의 첫날
- BM: 주말이 아닌 평일 중에서 각 달의 마지막 날
- BMS: 주말이 아닌 평일 중에서 각 달의 첫날
- WOM-2THU: 각 달의 두번째 목요일
- Q-JAN: 각 분기의 첫달의 마지막 날
- Q-DEC: 각 분기의 마지막 달의 마지막 날


```python
pd.date_range('2018-4-1','2018-4-30',freq='B') # 평일
pd.date_range('2018-1-1','2018-12-31',freq='W') # 일요일
pd.date_range('2018-1-1','2018-12-31',freq='B') # 월요일 등등 ~
```

```python
s = pd.Series(pd.date_range('2020-12-25',periods=100,freq='D'))
s.dt.year # 연도 추출
s.dt.weekday # 요일 추출
s.dt.strftime('%Y년 %m월 %d일')
```
