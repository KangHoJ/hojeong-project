```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family='NanumBarunGothic')
```

```python
rank = pd.read_excel('/content/drive/MyDrive/Miniproject_No_1/0.Code&Data/재승/행정구별랭킹.xlsx')
rank.columns
```

```python
rank
```

```python
rank["구별 치킨집 개수"] = rank["구별 치킨집 개수"].str.split().str[0]
```

```python
rank['면적 대비 폐업랭킹(치킨)'] = rank["면적 대비 폐업랭킹(치킨)"].str.split().str[0]
rank['가성비'] = rank["가성비"].str.split().str[0]
```

```python
rank2 = rank[['면적 대비 폐업랭킹(치킨)','순위']].set_index('면적 대비 폐업랭킹(치킨)').sort_index()
rank2
```

```python
chikenjib = rank[['구별 치킨집 개수',' 개수 순위']].set_index('구별 치킨집 개수').sort_index()[' 개수 순위']
gasungbi2 = rank[['가성비','가성비 순위']].set_index('가성비').sort_index()
```

```python
rank2['치킨집개수순위'] = chikenjib
rank2['가성비순위'] = gasungbi2
rank2.columns = ['면적 대비 폐업랭킹(치킨)','치킨집개수순위','가성비순위']
```

```python
rank2['치킨집개수순위']
```

```python
rank2['합계 점수(낮을 수록 유리)'] = rank2['면적 대비 폐업랭킹(치킨)'] + rank2['치킨집개수순위'] + rank2['가성비순위']
```

```python
rank2
```

```python
#rank2.set_index('행정구명')
rank2 = rank2.rename_axis(index='행정구명')
```

```python
rank2.columns = ['행정구명', '시군구 면적별 폐업률', '경쟁 치킨집 수', '임대료 대비 매출액', '합계 점수(낮을 수록 유리)']
```

```python
rank2
```

```python
rank2.set_index('행정구명')
```

```python
rank2 = rank2.sort_values('합계 점수(낮을 수록 유리)')
rank2.reset_index(drop=False,inplace=True)
```

```python
rank2
```

```python
plt.figure(figsize=(10,10))
plt.title('지역구 상권 선정')
ax = sns.barplot(y=rank2['행정구명'], x=rank2['합계 점수(낮을 수록 유리)'], dodge=False, palette='Spectral')
ax.tick_params(axis='y',labelsize=14)
for p in ax.patches:
    x, y, width, height = p.get_bbox().bounds
    ax.text(width*1.01, y+height/2, "%.0f"%(width), va='center')
plt.show()
```
