```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Miniproject_No_1/0.Code&Data/datas/주변시설/성동구 학교들.csv')
```

```python
df.head(5)
```

```python
df = df.loc[(df['도로명주소'].str.contains('성동구'))]
```

```python
df
```

```python
df['행정동'] = df['도로명주소'].str.split('(').str[1]
```

```python
df['행정동'] = df['행정동'].str.split(')').str[0]
```

```python
df['행정동'] = df['행정동'].str.split('.').str[0]
```

```python
df['행정동']
```

```python
df_graph = df['행정동'].value_counts()
df_graph
```

```python
plt.figure(figsize=(15, 5)) 
# 행정동별 초중고개수
colors = sns.color_palette('hls',20)
plt.bar(df_graph.index , df_graph.values ,color=colors)
plt.xlabel('행정동')
plt.ylabel('개수')
plt.title('행정동별 초,중,고 개수')
plt.savefig('행정동별 초,중,고 개수')
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
# 지도 시각화
import folium
from folium.plugins import MarkerCluster
from folium import Marker
import requests
```

```python
seoul_geojson = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
response = requests.get(seoul_geojson)
data_dict = response.json()
seongdonggu = {'type': 'FeatureCollection',
 'features': [feature for feature in data_dict['features'] if "성동구" in feature["properties"]["name"]]}
```

```python
school_map = folium.Map(
    location=[37.57,126.99],
    zoom_start=11)
folium.Choropleth(
    geo_data = seongdonggu,
    fill_color = '#22AA44',
    fill_opacity=0.4,
    line_opacity=1,
    school_map = folium.Map(location = [37.468251, 126.886212],
                          zoom_start = 20)
).add_to(school_map)
school_map
```

```python
for i in range(len(data_concat)):
    marker01 = folium.Marker([data_concat.loc[i]['위도'], data_concat.loc[i]['경도']],
                             icon = folium.Icon(color='blue'))
    marker01.add_to(school_map)
    
school_map
```
