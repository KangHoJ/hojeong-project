```python
top_bus_area.sort_values(ascending = False,inplace=True)
low_bus_area.sort_values(ascending = False,inplace=True)
top_bus.sort_values(ascending = False,inplace=True)
low_bus.sort_values(ascending = False,inplace=True)
```

```python
top_bus_area2 = top_bus_area.head(3)
low_bus_area2 = low_bus_area.tail(3)
top_bus2 = top_bus.head(3)
low_bus2 = low_bus.tail(3)
```

```python
top_bus_area3 = top_bus_area.head(25)
top_bus_area3
```

```python
top_low_bus_area= pd.concat([top_bus_area2,low_bus_area2]) 
top_low_bus= pd.concat([top_bus2,low_bus2])
```

```python
# 버스 정류장 개수(상위3개 하위3개)
colors = sns.color_palette('hls',6)
plt.bar(top_low_bus.index , top_low_bus.values ,color=colors)
plt.xlabel('시군구')
plt.ylabel('버스 정류장 개수')
plt.title('시군구별 버스 정류장 개수(상,하위 3개)')
plt.savefig('시군구별 버스 정류장 개수')
```

```python
# 면적대비 버스 정류장 개수(상위3개 하위3개)
colors = sns.color_palette('hls',6)
plt.barh(top_low_bus_area.index , top_low_bus_area.values ,color=colors)
plt.xlabel('시군구')
plt.ylabel('면적대비 버스 정류장 개수')
plt.title('시군구별 면적대비 버스 정류장 개수(상,하위 3개)')
plt.savefig('면적대비 버스 정류장 개수')
```
