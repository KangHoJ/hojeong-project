```python
school_raw = pd.read_excel('/content/drive/MyDrive/Miniproject_No_1/0.Code&Data/datas/주변시설/전국유치원,초중고_주소_2022.xlsx')
school_raw = school_raw.rename(columns=school_raw.iloc[2])

school_raw

school_df = school_raw[['학제', '시도', '행정구', '학교명',  '도로명주소']].dropna(axis=0)
school_df
```

```python
# 서울에 있는 초중고만 추출
seoul_school_df = school_df[school_df['시도'].str.contains('서울')]
seoul_school_df
seoul_school_lo = seoul_school_df.reset_index()
seoul_school_lo
```

```python
# 학교명과 주소만 남긴 다음에 주소를 위도 경도로 변환하기

seoul_school_address = seoul_school_lo[['도로명주소']]
seoul_school_address
```

```python
 seoul_school_address.info()
```

```python
import re

regex = "\(.*\)|\s-\s.*"

for i in range(len(seoul_school_address['도로명주소'])):
  seoul_school_address['도로명주소'][i] = re.sub(regex, '', seoul_school_address['도로명주소'][i])
  print(seoul_school_address['도로명주소'][i])
```

```python
seoul_school_address["도로명주소"] = seoul_school_address["도로명주소"].str.replace(pat=r'[^\w\s]', repl=r'', regex=True)
seoul_school_address.to_csv('seoul_school_place.csv')
```
