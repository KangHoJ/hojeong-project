```python
df = pd.read_csv('서울시 일반음식점 인허가 정보.csv')
df.head(3)
```

```python
data = df[['폐업일자','상세영업상태명','소재지면적','지번주소','업태구분명']]
data.head(3)
```

```python
data['시군구'] = data['지번주소'].str.split(' ').str[1]
data
```

```python
data['행정동'] = data['지번주소'].str.split(' ').str[2]
data
```

```python
data = data[['폐업일자','상세영업상태명','소재지면적','시군구','행정동','업태구분명']]
data.head(2)
```

```python
data2 = data.loc[(data['폐업일자'].str.contains('2023')) & (data['상세영업상태명']=='폐업')]
```
