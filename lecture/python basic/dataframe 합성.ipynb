## dataframe 합성

```python
df1 = pd.DataFrame({
    '고객번호':[1001,1002,1003,1004,1005,1006,1007],
    '이름':['둘리','도우너','또치','길동','희동','마이콜','영희']
}, columns=['고객번호','이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호':[1001,1001,1005,1006,1008,1001],
    '금액': [10000,20000,15000,5000,100000,30000]
},columns=['고객번호','금액'])
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='outer')
# 키 값이 한쪽에만 있어도 데이터를 보여줌 
```

```python
pd.merge(df1,df2,how='left') #왼쪽 df를 기준으로 합침
```

```python
pd.merge(df1,df2,how='right') #오른쪽 df를 기준으로 합침
```

```python
df1 = pd.DataFrame({
    '품종' :['setosa','setosa','virginica','virginica'],
    '꽃잎길이' :[1.4,1.3,1.5,1.3]},
    columns=['품종','꽃잎길이'])
df1
```

```python
df2 = pd.DataFrame({
    '품종' :['setosa','virginica','virginica','virginica'],
    '꽃잎길이' :[0.4,0.3,0.5,0.3]},
    columns=['품종','꽃잎길이'])
df2
```

```python
pd.merge(df1, df2)
```
